En algunos casos podemos representarlos de maneta gráfica y en otros casos podemos obtener información de forma analítica computando algúnos datos primordiales (promedios, medianas, etc, etc...) de nuestros datos que nos ayudan a generalizar y comenzar a obtener parámetros importantes sobre nuestro análisis.

#### _Histogramas._

Los histogramas son una representación gráfica del numero de veces que se lleva a cabo la sucutación de un evento. En otras palabras es la frecuencia con la que uno o varios datos se hace presente. 

Para ello es necesario trabajar con arreglos (**numpy**) y un administrador de estilos, por ejemplo **seabor**, quién da los diversos formatos al gráfico.

Para poder fijar, que durante todo nuestro análisis, los estilos sean administrados mediante **seaborn** será necesario ejecutar la función `sns.set()` durante la importación de las librerías correspondientes.

El primer histograma lo realizaremos con datos del arreglo nombrado `versicolor_petal_length`.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()

versicolor_petal_length=np.array([4.7, 4.5, 4.9, 4. , 4.6, 4.5, 4.7, 3.3, 4.6, 3.9, 3.5, 4.2, 4. ,4.7, 3.6, 4.4, 4.5, 4.1, 4.5, 3.9, 4.8, 4. ,
                                 4.9, 4.7, 4.3, 4.4, 4.8, 5. , 4.5, 3.5, 3.8, 3.7, 3.9, 5.1, 4.5, 4.5, 4.7, 4.4, 4.1,4. , 4.4, 4.6, 4. , 3.3,
                                 4.2, 4.2, 4.2, 4.3, 3. , 4.1])
```

Los resultados son alojados dentro de una variable _"domy"_ llamada (`_`) como parte de buenas prácticas: 

```python
_=plt.hist(versicolor_petal_length)
plt.show()
```

Una vez impreso el histograma vemos que necesitamos etiquetar a sus ejes. De lo contrario su interpretación puede provenir de diversas fuentes.

* `_=plt.ylabel('count')` para especificar que frecuencia.
* `_=plt.xlabel('petal length (cm)')` son las médidas en centímetros de los pétalos del arreglo `versicolor_petal_length`.

```python
_ = plt.hist(versicolor_petal_length)

_=plt.ylabel('count')
_=plt.xlabel('petal length (cm)')

plt.show()
```

Uno de los parámetros muy importantes es el del número de clases en un _histograma_ (`bins`). Es quién decide, mediante diferentes métodos, el cálculo del número de clases como; raíz cuadrada del número de datos ($\sqrt{n}$) `sqrt` y Sturges `sturges` por ejemplo.

Estos parámetros están dentro de la función `hist()`. Pero a manera de ejemplo, en esta ocasión, utilizaremos la raíz cuadrada del número de datos ($\sqrt{n}$) del arreglo `versicolor_petal_length`, mediante la función `len()` y el método de **numpy** `.sqrt()`, para depués pasarlo como parámetro `bin` a la función `hist()` para ser impreso:

```python
n_data=len(versicolor_petal_length)
n_bins=np.sqrt(n_data)
n_bins=int(n_bins)
_=plt.hist(versicolor_petal_length, bins=n_bins)

_ = plt.xlabel('petal length (cm)')
_ = plt.ylabel('count')

plt.show()
```

En el siguiente ejercicio graficaremos los datos del archivo `iris.csv` utilizando el estadístico **Bee swarm**. Para ello necesitamos nombrar las columnas dentro de la función `read_csv()`, de la librería de **pandas**, con el parámetro `names=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)','species']`, y añadiendo también `header=0`, como otro parámetro, para que no tome la observación 0 como título del _Dataframe_ `df`. 

```python
df=pd.read_csv('iris.csv', names=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)','species'], header=0)
df
```

Ahora es turno de la librería **matplotlib**. El gráfico _Bee swarm_ es invocado mediante el método `swarmplot`. Necesita como parámetro `x=` al atributo en el _Dataframe_ que contenga la variable cualitativa, en este caso es `scpecies`. Por otro lado necesita el atributo `y=`, que es el encargado de almacenar cualquier atributo cuantitativo en el _Dataframe_ que deseemos apreciar analizar.

Como un último parámetro necesitamos la fuente de los datos. El parámetro `data=` hará referencia al _Dataframe_ con los records para analizar. En este caso es `df`.

Para finalizar realizamos el etiquetado de los ejes, para poder concluir que los pétalos de _I. setosa_ son más cortos que los de _I. virginica_. También podemos observar que es muy raro encontrar petalos de _I. virginica_ más pequeños que los de _I. versicolor_.

```python
_=sns.swarmplot(x='species', y='petal length (cm)', data=df)
_ = plt.xlabel('species')
_ = plt.ylabel('petal length (cm)')
plt.show()
```
