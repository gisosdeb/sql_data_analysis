# NULL o IS NULL

En **SQL**, `NULL` representan records desconocidos o ignorados en los atributos de una tabla. Podemos investigar estos tipos de records usando la expresión `IS NULL`.

—_Investiguemos el número de valores `NULL` del atributo `birthdate` para la tabla `people`:—_

````python
import pandas as pd
import psycopg2.extras
conn = psycopg2.connect("dbname='test' user='test' host='/tmp/'")
cur = conn.cursor()
````

````python
cur = conn.cursor(cursor_factory=psycopg2.extras.DictCursor)
cur.execute(
    """
    SELECT COUNT(*) FROM intro_sql.people WHERE birthdate IS NULL;
    """
)
data = cur.fetchall()
conn.commit()
cur.close()
pd.DataFrame([i.copy() for i in data])
````

Por el contrario, si deseamos encontrar cuantos records o cuales elementos no son `NULL` podemos usar el _key word_ `IS NOT NULL`.

—_Averiguemos cuales elementos en el atributo `name`, de la tabla `people`, no son `NULL`:—_

````python
cur = conn.cursor(cursor_factory=psycopg2.extras.DictCursor)
cur.execute(
    """
    SELECT people.name FROM intro_sql.people WHERE birthdate IS NOT NULL;
    """
)
data = cur.fetchall()
conn.commit()
cur.close()
pd.DataFrame([i.copy() for i in data])
````

—_El nombre de las personas que siguen vivas en la tabla `people`. En otras palabras, aquellos records en el atributo `deathdate` que sea desconocidos o ignorados:—_

````python
cur = conn.cursor(cursor_factory=psycopg2.extras.DictCursor)
cur.execute(
    """
    SELECT people.name FROM intro_sql.people WHERE deathdate IS NULL;
    """
)
data = cur.fetchall()
conn.commit()
cur.close()
pd.DataFrame([i.copy() for i in data])
````

# LIKE o NOT LIKE

Ya hemos realizado filtros donde las consultas a las tablas tienen que ver más con atributos cuantitativos.  —_¿Pero qué pasa cuándo los atributos son cualitativos?...—_

En **SQL**, el operador `LIKE` puede ser utilizado en la clausula `WHERE` para buscar patrones en los records de los atributos. Pero para completar esta tarea necesitamos utilizar algo llamado _wilcard_.

En computación un _wilcard_ son simbolos que nos permiten representar uno o más carcteres. Por ejemplo: —_`%` nos ayuda a coincidir, en un texto, uno, varios o ningún caracter que nosotros hayamos proporcionado para comparar...—_

_—El siguiente query nos ayuda a localizar cualquier nombre de una compañia que coincida con `Data`.—_
```sql
SELECT name
FROM companies
WHERE name LIKE 'Data%';
```
Observamos que nos hemos apoyado de `WHERE` para realizar el filtro en el atributo `companies`, y `%` para buscar cualquier patron que coincidía con `Data`.

El _wildcard_ `_` se utiliza cara coincidir solamente un caracter. Por ejemplo: — _sabremos que nos referimos a `Estudiante` si solo digitamos como `Est_diante` o `Nissan` por `Nis_an`..._ —
```sql
SELECT name
FROM companies
WHERE name LIKE 'Nis_an';
```

````python
cur = conn.cursor(cursor_factory=psycopg2.extras.DictCursor)
cur.execute(
    """
    SELECT people.name FROM intro_sql.people WHERE name LIKE '_r%'
    """
)
data = cur.fetchall()
conn.commit()
cur.close()
pd.DataFrame([i.copy() for i in data])
````

`NOT LIKE` es la parte opuesta. Es decir, `NOT LIKE` se utiliza como negación o _"lo contrario a"_. Por ejemplo: — _si queremos encontrar todos los nombres de las personas que **"no"** comiencen con la letra `A`..._ —


````python
cur = conn.cursor(cursor_factory=psycopg2.extras.DictCursor)
cur.execute(
    """
    SELECT people.name FROM intro_sql.people WHERE name NOT LIKE 'A%'
    """
)
data = cur.fetchall()
conn.commit()
cur.close()
pd.DataFrame([i.copy() for i in data])
````
